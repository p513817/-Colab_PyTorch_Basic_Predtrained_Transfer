# 使用自己的圖片進行 inference

```
!python detect.py --source ../../images/my_cat --weights runs/exp1/weights/best.pt
```

```
from google.colab.patches import cv2_imshow
import cv2
import os

path = '../../images/my_cat/'
size = 400
ls_img =[]

for i,f in enumerate(os.listdir(path)):
  
  src = cv2.resize( cv2.imread(f'../../images/my_cat/{f}'), (size,size))
  res = cv2.resize( cv2.imread(f'inference/output/{f}'), (size,size))
  ls_img.append( cv2.hconcat([src, res]))  
  
  if i%2==1: cv2_imshow(cv2.hconcat([ls_img[i-1], ls_img[i]]))
  
```
